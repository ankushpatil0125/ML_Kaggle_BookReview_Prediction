## Import preprocessed data stored in csv file 

```python
path_link="/kaggle/input/test-dataset/final_preprocessed_test.csv"
df=pd.read_csv(path_link)
path_link="/kaggle/input/train-preprocessed/final_preprocessed.csv"
dft=pd.read_csv(path_link)
```

```python
df.head()
```

## Replace na with ' '

```python
df['review_text'].fillna('', inplace=True)
dft['review_text'].fillna('', inplace=True)
```

## Perform Tfidf Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,2))
X_train_tfidf = tfidf_vectorizer.fit_transform(dft['review_text'])
X_test_tfidf = tfidf_vectorizer.transform(df['review_text'])
```

## Import trained model and do prediction

```python
import pickle
pickled_model = pickle.load(open('/kaggle/input/testdata/trained_model.pkl', 'rb'))
df['rating']=pickled_model.predict(X_test_tfidf)
```

## make output file as per required format

```python
df.drop('review_text', axis=1, inplace=True)
```

```python
path_link="/kaggle/input/original-test/test.csv"
dfc=pd.read_csv(path_link)
df['review_id']=dfc['review_id']
```

```python
desired_order = ['review_id', 'rating']

# Change the order of columns in the DataFrame
df = df[desired_order]
```
